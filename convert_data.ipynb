```python
import csv
import numpy as np
import shutil
import os

filename = './score1/BITVehicle/VehicleInfo.csv'

category = {'Sedan': 0, 'SUV': 1, 'Microbus': 2, 'Minivan': 3, 'Bus': 4, 'Truck': 5} 

num = np.zeros(6)
with open(filename, "r") as csvfile:
    csvreader = csv.reader(csvfile)
    for pic in csvreader:
        if pic[0] != '':
            num[category[pic[8]]] += 1
    
print(num)
```

```python
info = np.zeros(5)
with open(filename, "r") as csvfile:
    csvreader = csv.reader(csvfile)
    for pic in csvreader:
        if pic[0] != '':
            filename = pic[1][0:-4]
            height = int(pic[2])
            width = int(pic[3])
            if int(pic[9]) == 1:
                info[0] = category[pic[8]]
                info[1] = (int(pic[4]) + int(pic[6])) / (2 * width)
                info[2] = (int(pic[5]) + int(pic[7])) / (2 * height)
                info[3] = (int(pic[6]) - int(pic[4])) / width
                info[4] = (int(pic[7]) - int(pic[5])) / height
                np.savetxt('./data/labels/' + filename + '.txt', info, fmt="%.6f", newline=' ')
            else:
                print("more than 1 vehicle in the image: ", filename)
                current_folder = "./data/images"  
                target_folder = "./data/multiimg"  
                filename += '.jpg'
                source_path = os.path.join(current_folder, filename)
                target_path = os.path.join(target_folder, filename)
                shutil.move(source_path, target_path)

```
